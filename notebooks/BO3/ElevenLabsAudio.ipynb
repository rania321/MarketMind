```python
import io
import os
import ipywidgets as widgets
from IPython.display import display, Audio
from elevenlabs import generate, set_api_key, Voice, VoiceSettings
from pydub import AudioSegment
from pydub.playback import play

# Set API Key
set_api_key("sk_924547070acfde9a9673ec1fbf1705c6139cd0c632da4962")  # Replace with your actual key

# Get available voices
from elevenlabs import voices
available_voices = {voice.name: voice.voice_id for voice in voices()}
print("Available voices:", available_voices)

# Tone settings
tone_settings = {
    "friendly": {
        "voice_id": available_voices["Aria"],
        "stability": 0.7,
        "similarity_boost": 0.75,
        "style": 0.5,
        "speaker_boost": True
    },
    "professional": {
        "voice_id": available_voices["Matilda"],
        "stability": 0.9,
        "similarity_boost": 0.85,
        "style": 0.3,
        "speaker_boost": True
    },
    "energetic": {
        "voice_id": available_voices["Laura"],
        "stability": 0.5,
        "similarity_boost": 0.8,
        "style": 0.7,
        "speaker_boost": True
    }
}

def generate_voiceover(text, tone="friendly"):
    params = tone_settings[tone]

    voice_settings = VoiceSettings(
        stability=params["stability"],
        similarity_boost=params["similarity_boost"],
        style=params["style"],
        speaker_boost=params["speaker_boost"]
    )

    voice = Voice(
        voice_id=params["voice_id"],
        settings=voice_settings
    )

    # Generate audio
    audio = generate(text=text, voice=voice)

    # Convert bytes to pydub AudioSegment
    return AudioSegment.from_mp3(io.BytesIO(audio))

def add_background_music(voiceover, music_path, volume_reduction=10):
    # Load music
    music = AudioSegment.from_file(music_path)

    # Trim or loop music to match voiceover length
    if len(music) < len(voiceover):
        loops = len(voiceover) // len(music) + 1
        music = music * loops  # Repeat music
    music = music[:len(voiceover)]  # Trim to exact length

    # Lower music volume and overlay
    music = music - volume_reduction
    combined = music.overlay(voiceover)

    # Fade-in & Fade-out for a smooth effect
    combined = combined.fade_in(2000).fade_out(2000)
    return combined

# UI Elements
tone_dropdown = widgets.Dropdown(
    options=["friendly", "professional", "energetic"],
    value="friendly",
    description="Tone:",
)

text_input = widgets.Text(
    value="Discover our new collection—crafted for comfort and style!",
    description="Text:",
    layout=widgets.Layout(width="80%"),
)

generate_button = widgets.Button(description="Generate Audio")
output_audio = widgets.Output()

def on_generate_button_clicked(b):
    with output_audio:
        output_audio.clear_output()
        print("Generating audio...")
        voiceover = generate_voiceover(text_input.value, tone_dropdown.value)

        # Add background music (make sure the file exists)
        bg_music_path = "/content/Nujabes - The Space Between Two World.mp3"  # Upload your music file
        if os.path.exists(bg_music_path):
            final_audio = add_background_music(voiceover, bg_music_path)
        else:
            final_audio = voiceover

        # Save and play
        final_audio.export("final_ad.mp3", format="mp3")
        play(final_audio)

        # Display audio player
        display(Audio("final_ad.mp3"))

generate_button.on_click(on_generate_button_clicked)

# Display UI
display(tone_dropdown, text_input, generate_button, output_audio)
```
